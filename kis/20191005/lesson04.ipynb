## 4.1 ユーザ定義関数  

同じコードを何度も再利用するときは、まとまった処理として関数化するとよいです。  

関数はこのように記述します。

引数の型と個数は自由です。戻り値も同じです。  
returnで結果を出力します。結果を出力しない場合はreturnがなくてもよいです。(何も返さない場合はpassを使います。) 



```
#何もしない関数
def do_nothing():
    pass
```

呼び出す時はこのように書きます。

```
do_nothing()
```

引数としてcolor  
戻り値として文字列を返す関数です。  

```
def FavoriteCity(city): # 仮引数
    return "I like " + city

output = FavoriteCity('Kumamoto') # 実引数
print(output)
```

誕生日を入力して、年齢を計算する関数の例です。  
引数をセットする順番を間違えないようにしてください。

```
from datetime import date # Pythonのライブラリーを使う

def cal_age_in_days(year, month, day):
    today = date.today() # 今日の日付
    birthday = date(year, month, day) # 誕生日
    age = today - birthday
    return age.days / 365
    
days = cal_age_in_days(2002, 10, 1) # 呼び出し
print('You are {0:.0f} days old.'.format(days))
days = cal_age_in_days(1989, 7, 19) # また呼び出し
print('You are {0:.0f} days old.'.format(days))
```

## 4.2 Noneについて

Pythonの特殊な値で、偽になるが、Falseと同じではない

```
print(None == False)
```

Noneは「空の値」と「存在しない値」を区別するために必要です。他の言語だとNULLにあたる。

```
def is_none(thing):
    if thing is None:
        print("It's None")
    elif thing:
        print("It's True")
    else:  
        print("It's False")
        
is_none(None)
is_none(0)
is_none([])
is_none([1])
```

関数に出力がない時に、return Noneを使うことがあります。

```
def get_price(fruit):
    fruitprice = dict([('apple', 150), ('orange', 100), ('banana', 50)])
    try:
        return fruitprice[fruit]
    except:
        return None # fruitのキーが見つからない時

def print_price(fruit):
    p = get_price(fruit)
    if p == None:
        print('No price found for {}'.format(fruit))
    else:
        print('Price of {} is {} yen'.format(fruit, p))

print_price('banana')
print_price('kiwi')
```

## 4.3 関数の引数について

引数が複数ある場合は順番が重要になってきます。

```
def birthday(year, month, day):
    return '{}年{}月{}日'.format(year, month, day)
    
print(birthday(2019, 10, 9))
print(birthday(9, 10, 2019)) 
```

引数にキーワードとしてセットするものを指定できるが、あまり使いません。コードも長くなってしまう欠点があります。

```
print(birthday(day=22, month=2, year=2002))
```

よく使うのは、呼び出す時ではなく、関数を定義するときに引数に対してデフォルト値を設定します。
関数を呼び出す時に引数が設定されなかったらデフォルト値を使って処理します。

```
def birthday(year=1900, month=1, day=1):
    return '{}年{}月{}日'.format(year, month, day)
    
print(birthday(2019, 10))
print(birthday(2019))
print(birthday(day=9, month=9))
```

## 4.4 引数のタプル化

アスタリスク\*を使うと、可変個の引数をタプルにまとめて仮引数にセットすることができます。
下記の場合、val1とval2は必須です。3個目以降の引数がタプル化されてまとめられます。

```
def print_more(val1, val2, *args): 
    print('First: ', val1)
    print('Second: ', val2)
    print('The rest: ', args)
    
print_more('tea', 'coffee')
print_more('tea', 'coffee', 'juice', 'beer', 'wine')
```

## 4.5 関数のdocstring  

関数の先頭にドキュメントを付けて関数の定義を表示することができます。


```
def print_if_not_empty(item):
    '''
    引数itemが空でない場合は
    print関数で要素を出力します。
    Input: item
    '''
    if item not in [None, [], (), {}]:
      print(item)

print_if_not_empty({})
print_if_not_empty([1, 2, 3])
```

help()関数でdocstringを呼び出すことが可能です。help関数も組み込み関数の1つです。

```
help(print_if_not_empty)
```

もちろん組み込み関数の説明もhelp関数で見ることができます。

```
help(print)
```

## 4.6 無名関数：lambda  


簡単に書ける関数に使用する無名関数です。
頻繁に繰り返し使う場合は名前を付けて関数定義しましょう。


```
def square(x):
    return x**2

# 二乗のリスト
sq = [square(x) for x in range(10)]
print(sq)
```

squareの関数をmap()関数を使ってリストにします。  
map関数はリストやタプルのすべての要素に同じ演算を適用する関数です。
ここでは、0～9までの数値に対してすべてsquare関数の計算を適応します。

```
def square(x):
    return x**2

sq = list(map(square, range(10)))
print(sq)

# lambdaを使うと簡潔になる
sq = list(map(lambda x: x**2, range(10)))
print(sq)

# 内包表記を使う場合はもっと簡単
sq = [x**2 for x in range(10)]
print(sq)
```

lambdaをfilter()関数を使うことで条件でフィルターすることができます。

```
# 偶数のリスト
even = list(filter(lambda x: x % 2 == 0, range(20)))
print(even)

# 内包表記を使う場合はもっと簡単
even = [x for x in range(20) if x % 2 == 0]
print(even)
```

※lambdaがもっとも役にたつのはソートする時のキーを設定したときです。

```
# Name, grade, age
students = [('Abe', 'A', 20), ('Tanaka', 'B', 18), ('Suzuki', 'B', 19)]
# sort by name
students.sort(key=lambda x: x[0])
print(students)
# sort by grade, age
students.sort(key=lambda x: (x[1], x[2]))
print(students)
# sort by age
students.sort(key=lambda x: x[2])
print(students)
```
