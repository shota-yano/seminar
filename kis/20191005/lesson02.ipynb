## 2.3 条件分岐(if文)

条件によって処理を分けたいときに使用します。
条件にはTrueかFalseになるような書き方をします。
条件の式の最後はコロンが必要です。

※注意
分岐したときの処理はインデントを4スペース入れてから書いてください。タブを1つ入れても良いです。ただし、プログラム内でインデントはスペースとタブが混在しないようにしてください

例：値が＋か-かを判断する場合です。

```
a = 100

print('a>0の判定結果は',a>0)

if a > 0:
    print('aは正の数です')
else:
    print('aは負の数です')
```

公式では4スペースだが、2文字文でも可。

```
a = -10
if a > 0:
  print('aは正の数です')
else:
  print('aは負の数です')
```

複数の条件を書く場合はelifを使います。
また、等価であることの比較は==を使います。

比較演算子の例：

* a == 0  :aの値が0と等しい
* a != 0  :aの値が0ではない
* a >= 0  :aの値が0以上(0は含む)
* a > 0   :aの値が0より大きい(0は含まない)
* a <= 0  :aの値が0以下(0を含む)
* a < 0   :aの値が0より小さい(0は含まない)

```
a = 0
if a == 0:
    print('aは0です。')
elif a > 0:
    print('aは正の数です')
else:
    print('aは負の数です')
```


条件分岐により代入する値を変えることも可能です。  
例：aの値が0より大きい場合は0、小さい場合は1がセットされる

```
a = 10
b = 0 if a > 0 else 1
print(b)
```

論理演算子を使ってこのように条件分岐することもできます

論理演算子の例：
* A and B :AとBがTrueの時にTrueとなる
* A or B  :AかBのどちらかがTrueならTrue
* not A   :AがTrueでない

- 要素が入っている：in

```
a = 1
b = 2

if a > 0 and b > 0:
  print('aもbも正')
else:
  print('それ以外')
```

```
a = -1
b = 2

if a > 0 or b > 0:
  print('少なくともaかbのどちらかが正')
else:
  print('それ以外')
```

```
a = -1

if not a > 0 :
  print('aが0より小さい')
else:
  print('それ以外')
```

指定した要素のどれかと一致していればTrueとしたい場合はinを使います。

```
a = 0

if a in [0,1,2] :
  print('aは0か1か2')
else:
  print('それ以外')
```

Pythonでは一行のコードが長いとき、可読性を高めるためにコードを複数行に分けて書くことも可能です。  
継続文字のバックスラッシュ \ を使うことで改行を入れることが可能です。

```
#このように長い条件だと読みにくい
a, b, c, d = 3, -10, 1, 2
if (a > 0 or b > 0) and not c == 0 and d in [1,2,3]:
    print('Yes')
```

```
# バックスラッシュを使って分ける
if (a > 0 or b > 0) and \
    not c == 0 and \
    d in [1,2,3]:
    print('Yes')
```


## 2.4 反復処理

繰り返し処理したい場合はwhile文かfor分を使います。  

while文は条件がTrueになっている限り処理を実行します。  
※コロンを忘れないように注意してください。

```
i = 0
while i < 10:
    print(i)
    i += 1
```

for文はデータの集まりから1つずつ順番に取り出して、要素を全て取り出したら終了します。  

for データ in データの集まり

```
collection = [0, 1, 2, 3, 4]
for item in collection:
    print(item)

#このように書いてもよいです。
for item in range(5):
  print(item)
```

このようにデータの中から複数の要素を同時に取り出すことも可能です。

```
address = [('Tanaka','Kumamoto'), ('Suzuki','Fukuoka'), ('Sato','Kagoshima')]
for name, city in address:
    print('Name: {}, City: {}'.format(name, city))
```

## 2.5 breakとcontinue

ループ処理を途中でやめたい場合や特定の場所以降の処理をスキップさせたい場合に使用します。
* break   :ループ処理を終了する
* continue:ループ処理を次の要素に回す

```
#1が入力されるまで無限に実行する
while True:
    value = input("修了するには[1]を入力してください: ")
    if value == '1': # 終了
        break
    else:
        try: # エラー処理
            number = int(value)
        except: # 整数に変更できない場合
            print('{} は数字ではない'.format(value))
            continue # 次に回す
            
        print('入力した数字: {}'.format(value))
```
