
## 3.2 str.format関数

ここまででも時々登場していたものです。  
printで出力の時、formatを使って、引数の値を文字列に代入できます。  


```
a, b = 2 , 3
print('{} + {} = {}'.format(a, b, a + b))
print('{1} + {0} = {2}'.format(a, b, a + b))
```

変数がリストやタプルなどの場合は下記のように要素を指定することが出来ます
* {0[0]}であれば、0番目の引数の0番目の要素を指します

```
fruit1 = ('みかん', 80)
fruit2 = ('バナナ', 50)
fruit3 = ('りんご', 100)
str3 = '{0[0]}一つ{0[1]}円と{1[0]}一つ{1[1]}円、合わせて{2}円。'

total1 = str3.format(fruit2, fruit1, fruit2[1]+fruit1[1])
print(total1)

total2 = str3.format(fruit2, fruit3, fruit2[1]+fruit3[1])
print(total2)
```

浮動小数点数型に小数点以下の桁数をいくつ見せるか指定することもできます。
※デフォルトは小数点後6つ  
* f :浮動小数点のこと

今回は扱いませんが、f以外にもいくつかあります。  

```
pi = 3.1415926535
print('pi = {0:f}'.format(pi))
print('pi = {0:.3f}'.format(pi))
print('pi = {0:.8f}'.format(pi))
print('pi = {0:.0f}'.format(pi))
```

## 3.3 format関数

同じように文字列に対して整形するための関数です。
これは0や\*で埋める例です。


```
s = format(255, '08d')
print(s)
print(format('center', '*^16'))
```

```
#余談ですが0埋めについては別の方法もあります
s = '255'
s_zero = s.zfill(8)
print(s_zero)
```

## 3.4 split関数

文字列を単語ごとに分けてリストを返します。  
引数に何も指定しない場合はスペースで区切ります。特定の区切りたい場合はその文字を指定します。

```
str1 = 'I am \tlearning Python.\n'
print(str1.split())     #スペース区切り
print(str1.split('a'))  #aという文字で区切る
print(str1.split('\t')) #タブで区切る
```

## 3.6 append関数 

ここからリスト内関数を学びます。  
appendはリストの末尾に新しい要素を追加します。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
cities.append('Okinawa') # 末尾へ追加
print(cities)
```

## 3.7 extend関数  

リストの末尾にリストに含まれる要素を追加します。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
add_cities = ['Kochi', 'Ehime', 'Tokushima','Kagawa'] 
cities.extend(add_cities) # リストの結合、cities += a_few_moreもできる
print(cities)
```

extend関数以外だとこんな書き方でも大丈夫です。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
add_cities = ['Kochi', 'Ehime', 'Tokushima','Kagawa'] 
cities += add_cities
print(cities)
```

## 3.8 insert関数

## 3.10 pop関数  
リストの要素を1つ取り出して、リストから削除します。  
引数を指定しない場合は末尾から1つ取り出します。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
a = cities.pop() 
print(cities)
print(a)
```

引数を指定するとその要素を取り出すことができます。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
a = cities.pop(1) 
print(a)
```

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
a = cities.pop(-1) # 要素を取り出して、削除する、pop(1)やpop(-1)も指定できる
print(cities)
print(a)
```

## 3.14copy関数  
リストをコピーします。  
下記の場合、コピーさらたデータがセットされた b は a とは別のインスタンスなので a が変更されても b は影響を受けません。

```
a = [1, 2, 3]
b = a.copy()
a[1] = 9
print(a)
print(b)
```

copy関数を使わずにこのようにコードを書いた場合は注意が必要です。  
この場合、a で行った変更が b にも影響を与えます。

```
a = [1, 2, 3]
print(a)
b = a
print(b)
a[0] = 'x'
print(a)
print(b)
b[1] = 'y'
print(a)
print(b)
```

## 3.15 sort関数  
リストの中の要素を並び替えるときに使用します。実行後はリスト内が更新されます。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
cities.sort()
print(cities)

#降順にしたいとき
cities.sort(reverse=True)
print(cities)
```

## 3.16 sorted関数

リスト内の要素をソートします。ただし、先に紹介したsort関数と違いリストの中は更新されません。  
使い方の違いとしては
* sorted(ソートしたいリスト)
* ソートしたいリスト.sort()


```
wordlist =  ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
numlist = [2, 5, 3, 15, 9, 1, 7, 20, 10, 8]

# sorted関数でリストをソートして新しいリストを返す
sorted_wl = sorted(wordlist)
print('new sorted list', sorted_wl)
sorted_nl = sorted(numlist)
print('new sorted list', sorted_nl)

# もともとのリストは変わらない
print('original', wordlist)
print('original', numlist)

# リスト内のメソッドのsortではリスト内のリストをソートする
wordlist.sort()
numlist.sort()
print('list.sort', wordlist)
print('list.sort', numlist)
```

## 3.17 range関数  
引数に指定された数のリストを生成します。

```
a = range(10)
print(type(a))
print(a)
```

range関数はそのままだとこのようになるため、リストを返すためにはこのようにリストに変換します。  
リストの要素は0から始まることに注意してください。

```
a = list(range(10))
print(a)
```

1から始めたい場合はこのように書きます。

もし逆順にする場合はこのように書きます。

```
a = list(range(10,0,-1))
print(a)
```

```
for i in range(5):
    print(i)
```

step(増分)を指定することも可能です。  
下記の場合、1～9の整数の中から、2つ飛ばしでリストを作成します。

```
for i in range(1, 10, 2): # +2ずつ
    print(i)
```

## 3.18 スライス

pop関数以外にもスライスを行うことで、リストから要素を取り出すことができます。  
この時、pop関数と違って元のリストから値は削除されません。

```
cities = ['Kumamoto', 'Fukuoka', 'Saga', 'Nagasaki', 'Oita','Miyazaki','Kagoshima']
print(cities[2])  #インデックス2の要素を返す
print(cities[-1]) #一番最後のインデックスの要素を返す
print(cities[1:3])#インデックス1から3の手前までの要素を返す
print(cities[:3]) #インデックスが3より小さい要素を返す
print(cities[3:]) #インデックス3から最後の要素まで返す
print(cities[-3:]) #インデックスの最後から３つの要素を返す
print(cities[::2]) #最初から最後まで2つ飛ばしで要素を返す
print(cities[1:5:2]) #インデックス1から5より小さい要素の中で2つ飛ばしで要素を返す
```

文字列の場合はこうなります。

```
str2 = 'Today I study python at KIS'
print(str2[2])
print(str2[-1]) # 後ろから
print(str2[8:10])
print(str2[:7])
print(str2[11:])
print(str2[-3:]) # 最後の３文字
print(str2[::2]) # step=2
len(str2)
```

## 3.19 リスト内包表記

Pythonの特有の構文でよく見かけると思いますのでここで覚えておきましょう

ここでは、空のリストに0～9までの値を1つずつリストに追加していきます。

```
#今までの書き方
seq = []
for i in range(10):
    seq.append(i) 
print(seq)
```

リスト内包表記を使用するとこのように書けます。

```
seq = [i for i in range(10)]
print(seq)
```

if文を使って、リストの中から条件に一致する値だけ返すことが可能です。
例えばこの場合は偶数のみが残ります。

```
seq = [i for i in range(10) if i%2 == 0]
print(seq)
```

このようにすると偶数の場合はそのまま値を返し、それ以外は'奇数'という文字を返してリストを作ります。

```
seq = [i if i%2 == 0 else '奇数' for i in range(10)]
print(seq)
```

計算式を使うことができます。この場合各値が2乗されます。

```
seq = [i*i for i in range(10)]
print(seq)
```

取り出した値に関数を使うことも可能です。この場合絶対値のリストが作られます。

```
# 絶対値を返す
seq = [abs(i) for i in range(-5, 5)]
print(seq)
```

zip関数を使うと2つのリストを同時に使用することが可能です。

```
# 2つのリスト
list1 = [1, 2, 3]
list2 = [4, 5, 6]
ziplist = list(zip(list1, list2))
print(ziplist)
```

list1の値をxに代入し、list2をyに代入し、x * y を計算した結果をリストに格納します。

```
list1 = [1, 2, 3]
list2 = [4, 5, 6]
seq = [x * y for x, y in zip(list1, list2)]
print(seq)
```

タプル型を使って辞書型を生成することもできます。

```
pairs = [(1, 'a'), (2, 'b'), (3, 'c')]
dic1 = {x: y for (x, y) in pairs}
print(dic1)
```

２つのリストからzip関数で辞書型を生成することもできます。

```
keys = [1, 2, 3]
values = ['a', 'b', 'c']
dic1 = {x: y for (x, y) in zip(keys, values)}
print(dic1)
```

※補足：辞書型はenumerateを使って、キーを振ることもできます。

```
values = ['a', 'b', 'c']
dic1 = dict(enumerate(values))
print(dic1)
```

## 3.21 keys関数  
dict内のキーを取り出します。

```
dic1 = {0: 'a', 1: 'b', 2: 'c'}
keys = dic1.keys()
print(keys)
for i in keys:
    print('key: {} value: {}'.format(i, dic1[i]))
```

これはすべて同じ結果になります。

```
dic1 = {0: 'a', 1: 'b', 2: 'c'}
pairs = []
for i in dic1.keys():
    pairs.append((i, dic1[i]))
print(pairs)

# 内包表記を使って
pairs = [(i, dic1[i]) for i in dic1.keys()]
print(pairs)

# dictの関数を使って
pairs = list(dic1.items())
print(pairs)
```

## 補足  


辞書型に新しい値を入れたい場合はこのように書きます。

```
dic1 = {0: 'a', 1: 'b', 2: 'c'}
dic1[5] = 'x'
print(dic1)
```

dict型のキーはタプル型でもよいです。このような使い方をします。

```
# 誰がどんな部屋に住んでいる
block_num = [(5, 101), (3, 102), (7, 302)] # 棟と部屋番号
family_member = [['Amy', 'Jane', 'Mary'], ['John', 'Mike'], ['Charles']] # 住民
address = dict(zip(block_num, family_member))
print(address)

# もう一人が(7, 302)に引っ越してきたら
address[(7, 302)].append('Bob')
print(address)
```
