## 1.3 変数について  

変数は、データ(値)にラベル(名前)を付けて、後で使用できるようにするために使います。  
例えば、「2」という値に「a」という名前を付けてとっておきます。  
※「a」という変数に「2」という値を代入するとも言います。  

* a という変数に 2 という値を代入する  
* b という変数に 3 という値を代入する  


```python
a = 2
b = 3
a * b
```

a * b を実行したら、aとbに代入された値を使って計算します。そのため結果は6となります。  
変数はメモリ上に格納されているためこのように別のセルで呼び出して使用することが可能です。

```python
c = 10
a * c
```

同じ名前で変数が使われると、保持している値は上書きされるので注意してください。
試しにしたのセルを実行した後、上のセルを実行してみてください。

```python
a = 4
```

変数名は大文字小文字区別しますので注意してください

```python
a = 4
A = 2
print(a)
print(A)
```

## 1.4 データ型(コレクション)

- リスト型 (list)

```python
a = [1, 2, 3, 4]  
type(a)
```

構成部分は何でもいい。違う型が混ざってもいい。

```python
a = [1, 'abc', b'a', True]  
type(a)
```

内部の値は変更可能で要素の追加などの操作が可能

```python
a.append(99)
print(a)
```

リストの中にリストを入れることも可能

```python
a.append(['KIS','オープンセミナー'])
print(a)
```

機械学習とかを始めるとリストなどのコレクションは非常によく使う型なので使いこなせるようになりましょう。  

* タプル型 (tuple)

```python
b = (1, 'Name', 'Address')
type(b)
```

リストと違って、普通に構造を持っている。一度構築したら内部の値は変更不能で、リストのように要素を追加したりは出来ない。

* 集合型 (set/frozenset)

```python
c = {1, 2, 3}  
type(c)
```

重複の値を許さない。構築の時重複の値は消される。

```python
c = set([1, 2, 3, 3, 4])
c
```

{1, 2, 3, 4}だけ残る。setの内部の値は変更可能です。  

frozensetは変更不能なので、値が変わらない定数の保持などに使うとよいです。

```python
c = frozenset([1, 2, 3, 3, 4])
type(c)
```

* 辞書型 (dict)
  * Key-Valueというkyeと値のセットで保持します。
  * Keyと値は方が違ってもよいです。
  * 値はリスト型をセットしてもよいです。

```python
d =  {1:'Apple', 2:'Orange', 3:['Banana','Lemon']}
type(d)
```

※Keyのほうはリスト型はダメですが、タプル型は使用可能です。

```python
d2 = {[1,2]:'Apple', 3:'Orange', 4:['Banana','Lemon']}
```

```python
d2 = {(1,2):'Apple', 3:'Orange', 4:['Banana','Lemon']}
print(d2)
```

## 1.5 コメントアウト

プログラム中にコメントを入れたい場合は#を入れてください。
\#が入ると、それ以降の部分は処理対象外となります。  
これは1行のみ有効です。


```python
strName = 'KIS' #これより右は処理されない

# ここは処理されない
# strName = 'KIS Co.,Ltd'

print(strName)
```

複数行にわたってコメントアウトしたい場合はダブルクォート3つ使って囲います。

```python
strName = 'KIS' 

"""ダブルクォートを3連続で入れることで
このように複数行まとめてコメントアウトすることが可能です。
これはプログラムの先頭で関数の説明を入れるときに良く使われます
"""

print(strName)
```

## 1.6 出力

print()関数

変数や出力などを確認するのによく使います

```python
print(100)
print('あいうえお')
a = 2 + 7
print(a)
b = [1,2,3, 4, 5]
print(b)
```

複数の変数を一行で値を代入することもできます。

```python
a, b, c = 1, 2, [3, 4, 5]
print(a)
print(b)
print(c)
print(a, b, c)
```

formatを使うと、変数の値を文字列に代入できます。{}がフォーマットで指定した値に変わります。  
{}の中に引数を設定することで、引数の順序を指定して変数の代入場所を決めることが出来ます。  
※引数は0から数えます

```python
print('2 + 7 = {}'.format(2+7))
print('{1} + {0} = {2}'.format(a, b, a + b))
```

## 1.7 入力

input()関数  
ユーザーから入力を求めたい時に使います。
* 変数名 = input(‘文字列’)  

とすることで、ユーザが入力した値を受け取れます。  

```python
name = input("名前を入力してください ")
print("あなたの名前は {}".format(name))
type(name) # str
```

数値を入力しても文字として返されます。

```python
age = input("年齢を入力してください ")
print("あなたは {} 歳です。".format(age))
type(age)
```

型は文字列なので、そのままでは計算には使えません。
計算に使いたい場合は
* int()関数
* float()関数　　

などを使って型を変更する必要があります。

```python
age = int(age)
type(age) 
```

※変換後の型が扱えないものはエラーになります。

```python
int('a')       #数字ではないのでNG
int('1.5')     #intは小数を扱えないのでNG(floatならOK)
float('1.2.3') #ピリオドが2個あるのでNG
```
